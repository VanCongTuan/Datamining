```python
! pip install basemap
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

df = pd.read_excel("/content/drive/MyDrive/KPDL/BTL/lon & Lat.xlsx")
df_north_america = df[df["region"] == 'North America']
df_oceania = df[df["region"] == 'Oceania']
df_latin_america = df[df["region"] == 'Latin America']
df_europe = df[df["region"] == 'Europe']
df_africa = df[df["region"] == 'Africa']
df_asia = df[df["region"] == 'Asia']
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Tạo figure và axes cho bản đồ
fig, ax = plt.subplots(figsize=(12, 8))

# Tạo bản đồ thế giới
m = Basemap(ax=ax, projection='mill', lon_0=0)

# Vẽ các biên châu lục
m.drawcoastlines(linewidth=0.5)
# Vẽ biên giới của các lục địa
m.drawcountries(linewidth=0.5)

# Ghi chú tên các châu lục
continents = {
    'North America': (-100, 45),
    'Latin America': (-60, -20),
    'Europe': (15, 55),
    'Africa': (15, 10),
    'Asia': (95, 55),
    'Oceania': (145, -25),
}
for continent, (lon, lat) in continents.items():
    x, y = m(lon, lat)
    plt.text(x, y, continent, ha='center', va='center', fontsize=13, fontweight='bold')

# Chuyển đổi lon & lat sang tọa độ x, y trên bản đồ
x, y = m(df_north_america['Lon'].values, df_north_america['Lat'].values)

# Vẽ các điểm dựa trên tọa độ x, y
m.scatter(x, y, marker='o', color='b', zorder=5, s=25)

# Chuyển đổi lon & lat sang tọa độ x, y trên bản đồ
x, y = m(df_asia['Lon'].values, df_asia['Lat'].values)

# Vẽ các điểm dựa trên tọa độ x, y
m.scatter(x, y, marker='o', color='r', zorder=5, s=25)

# Chuyển đổi lon & lat sang tọa độ x, y trên bản đồ
x, y = m(df_europe['Lon'].values, df_europe['Lat'].values)

# Vẽ các điểm dựa trên tọa độ x, y
m.scatter(x, y, marker='o', color='yellow', zorder=5, s=25)

# Chuyển đổi lon & lat sang tọa độ x, y trên bản đồ
x, y = m(df_latin_america['Lon'].values, df_latin_america['Lat'].values)

# Vẽ các điểm dựa trên tọa độ x, y
m.scatter(x, y, marker='o', color='g', zorder=5, s=25)

# Chuyển đổi lon & lat sang tọa độ x, y trên bản đồ
x, y = m(df_africa['Lon'].values, df_africa['Lat'].values)

# Vẽ các điểm dựa trên tọa độ x, y
m.scatter(x, y, marker='o', color='Cyan', zorder=5, s=25)

# Chuyển đổi lon & lat sang tọa độ x, y trên bản đồ
x, y = m(df_oceania['Lon'].values, df_oceania['Lat'].values)

# Vẽ các điểm dựa trên tọa độ x, y
m.scatter(x, y, marker='o', color='pink', zorder=5, s=25)

# Hiển thị bản đồ
plt.title('Map of World')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
# Đường dẫn đến tệp Excel chứa tọa độ điểm
excel_file_path = '/content/drive/MyDrive/KPDL/BTL/Pak2.xlsx'

# Đọc dữ liệu từ tệp Excel
data = pd.read_excel(excel_file_path)
data_Balochistan = data[data["Province"] == 'Balochistan']
data_KP = data[data["Province"] == 'Khyber Pakhtunkhwa']
data_Punjab = data[data["Province"] == 'Punjab']
data_Sindh = data[data["Province"] == 'Sindh']
data_GB = data[data["Province"] == 'Gilgit-Baltistan']
data_Azad_Kashmir = data[data["Province"] == 'Azad Kashmir']
data_ICT = data[data["Province"] == 'Islamabad Capital Territory']

# Đường dẫn đến tệp Shapefile của Pakistan
shapefile_path = '/content/drive/MyDrive/KPDL/BTL/pak_adm_wfp_20220909_shp/pak_admbnda_adm1_wfp_20220909.shp'

# Đọc dữ liệu hành chính của Pakistan
data = gpd.read_file(shapefile_path)

# Tạo plot và đặt giới hạn bản đồ
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(60, 80)  # Điều chỉnh giới hạn dọc (longitude)
ax.set_ylim(22, 37)  # Điều chỉnh giới hạn ngang (latitude)

# Vẽ bản đồ hành chính của Pakistan
data.plot(ax=ax, edgecolor='black', linewidth=0.5, facecolor='lightgray')

# Vẽ các điểm trên bản đồ
ax.scatter(data_Balochistan['Lon'], data_Balochistan['Lat'], color='red', label='University of Balochistan', s=50)
ax.scatter(data_KP['Lon'], data_KP['Lat'], color='orange', label='University of Khyber Pakhtunkhwa', s=50)
ax.scatter(data_Punjab['Lon'], data_Punjab['Lat'], color='blue', label='University of Punjab', s=50)
ax.scatter(data_Sindh['Lon'], data_Sindh['Lat'], color='green', label='University of Sindh', s=50)
ax.scatter(data_GB['Lon'], data_GB['Lat'], color='yellow', label='University of Gilgit-Baltistan', s=50)
ax.scatter(data_Azad_Kashmir['Lon'], data_Azad_Kashmir['Lat'], color='pink', label='University of Azad Kashmir', s=50)
ax.scatter(data_ICT['Lon'], data_ICT['Lat'], color='BRown', label='University of Islamabad Capital Territory', s=50)


# Hiển thị chú thích
ax.legend()
# Hiển thị tên tỉnh và tọa độ
province_names = {
    'Balochistan': (64.0, 28.0),
    'Khyber'+'\nPakhtunkhwa': (72.9, 35.0),
    'Punjab': (73.0, 31.0),
    'Sindh': (68.0, 25.5),
    'Gilgit-Baltistan': (74.5, 35.8),
    'AJK': (73.9, 33.3),
    'ISLAMABAD': (74, 34)
}
for province, coords in province_names.items():
    x, y = coords
    ax.text(x, y, province, fontsize=8, ha='center')

plt.show()
```
